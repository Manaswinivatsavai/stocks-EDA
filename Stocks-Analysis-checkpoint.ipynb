```python
#Which stock showed the highest percentage change (% Chng) on the latest trading day
df["% Chng"].max()
print(f"{df.loc[df["% Chng"].idxmax(),"Symbol"]} Showed Highest stock : {df["% Chng"].max()}%")
#the average daily volume traded for all stocks
print(f"Average Daily Volume       : {df["Volume (lacs)"].mean():.2f}")
# stocks which have significantly higher or lower volume
print(f"Stocks with Higher Volume  :{df.loc[df["Volume (lacs)"].idxmax(),"Symbol"]} ({df["Volume (lacs)"].max()} lacs) ")
print(f"Stocks with Lower  Volume  :{df.loc[df["Volume (lacs)"].idxmin(),"Symbol"]} ({df["Volume (lacs)"].min()} lacs) ")
```

```python
#stocks have the highest and lowest 52-week high  and 52-week low prices
print(f"52-week high : {df['52w H'].max()}")
print(f"52-week low  :{df['52w H'].min()}")
```

```python
#correlation between ‘Turnover (crs.)’ and ‘Volume (lacs)’
df["Turnover (crs.)"]=df["Turnover (crs.)"].str.replace(",","")
cor=df["Turnover (crs.)"].corr(df['Volume (lacs)'])
print(f"Correlation is {cor:.2f}--->If turnover increases then volume increases")
df[["Turnover (crs.)","Volume (lacs)"]].corr()
```

```python
#How many stocks have positive vs negative ‘365 d % chng’? Which stocks performed best and worst over the past year?
print(f"Total Positive Stocks :{(df["365 d % chng"]>0).sum()}")
print(f"Toatl Negative Stocks :{(df["365 d % chng"]<0).sum()}")
print(f"{df.loc[df["365 d % chng"].idxmax(),"Symbol"]} performed Best  : {df["365 d % chng"].max()}%")
print(f"{df.loc[df["365 d % chng"].idxmin(),"Symbol"]} performed Worst : {df["365 d % chng"].min()}%")
```

```python
df.columns
```

```python
#stock having the highest turnover
df['Turnover (crs.)'] = df['Turnover (crs.)'].replace(',', '', regex=True).astype(float)
highest_turnover = df.loc[df['Turnover (crs.)'].idxmax()]
print(highest_turnover)
```

```python
#Top 5 most actively traded stocks (by volume)
df['Volume (lacs)'] = df['Volume (lacs)'].replace(',', '', regex=True).astype(float)
top_volume = df.sort_values(by='Volume (lacs)', ascending=False).head(5)
top_volume
```

```python
#Top 5 long-term gainers 
df['365 d % chng'] = df['365 d % chng'].replace('%', '').astype(float)
top_yearly = df.sort_values(by='365 d % chng', ascending=False).head(5)
top_yearly
```

```python
#Which stock had widest price range today
df['High'] = df['High'].replace(',', '', regex=True).astype(float)
df['Low'] = df['Low'].replace(',', '', regex=True).astype(float)
df['Range'] = df['High'] - df['Low']
most_volatile = df.sort_values(by='Range', ascending=False).head(5)
most_volatile
```
